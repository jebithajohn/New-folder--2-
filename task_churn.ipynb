```python
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels for object columns
train_df['Surname']= label_encoder.fit_transform(train_df['Surname'])
train_df['Surname'].unique()
train_df['Geography']= label_encoder.fit_transform(train_df['Geography'])
train_df['Geography'].unique()
train_df['Gender']= label_encoder.fit_transform(train_df['Gender'])
train_df['Gender'].unique()
test_df['Surname']= label_encoder.fit_transform(test_df['Surname'])
test_df['Surname'].unique()
test_df['Geography']= label_encoder.fit_transform(test_df['Geography'])
test_df['Geography'].unique()
test_df['Gender']= label_encoder.fit_transform(test_df['Gender'])
test_df['Gender'].unique()
```

```python

# splitting the dataset
from sklearn.model_selection import train_test_split
 
predictors = train_df.drop(['Exited','Surname','RowNumber','CustomerId','HasCrCard'], axis=1)

target = train_df["Exited"]
X_train1, X_test, y_train1, y_test = train_test_split(    predictors, target, test_size=0.2, random_state=0)
```

```python
print("Before OverSampling, counts of label '1': {}".format(sum(y_train1 == 1)))
print("Before OverSampling, counts of label '0': {} \n".format(sum(y_train1 == 0)))

# import SMOTE module from imblearn library
# pip install imblearn (if you don't have imblearn in your system)
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state = 2)
X_train, y_train = sm.fit_resample(X_train1, y_train1.ravel())

print('After OverSampling, the shape of X_train: {}'.format(X_train.shape))
print('After OverSampling, the shape of y_train: {} \n'.format(y_train.shape))

print("After OverSampling, counts of label '1': {}".format(sum(y_train == 1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train == 0)))
```

```python
#logistic regression
from sklearn import linear_model
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
log_param= {'penalty':['l1', 'l2', 'elasticnet', 'none'],'dual':[0,1],'solver':['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']}
scoring= ['accuracy']
log_randm_src = RandomizedSearchCV(estimator=linear_model.LogisticRegression(), param_distributions = log_param,cv = 2,scoring='accuracy', n_iter = 10, n_jobs=-1,verbose=1)
log_randm_src.fit(X_train, y_train)
log_best= log_randm_src.best_score_
```

```python
#decision tree classifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
parameters = {'criterion': [ 'gini', 'entropy', 'log_loss'],'max_depth'    : [1,4,5] ,'random_state': [1,5,3]               }
scoring= ['accuracy']
dec_randm_src = RandomizedSearchCV(estimator=model, param_distributions = parameters,scoring='accuracy',cv = 2, n_iter = 10, n_jobs=-1,verbose=1)
dec_randm_src.fit(X_train, y_train)
dec_best=  dec_randm_src.best_score_
```

```python
#random forest classifier
from sklearn.ensemble import RandomForestClassifier
regr = RandomForestClassifier(random_state=0, n_jobs=-1)


parameters = {'n_estimators':[10,50,100],'criterion': [ 'gini', 'entropy', 'log_loss'],'max_depth'    : [1,4,5] ,'random_state': [0,5,3] ,'max_features':['sqrt', 'log2', None]}            
scoring= ['accuracy']
ran_randm_src = RandomizedSearchCV(estimator=RandomForestClassifier(), param_distributions = parameters,scoring='accuracy',cv = 2, n_iter = 10, n_jobs=-1,verbose=1)
ran_randm_src.fit(X_train, y_train)

random_best= ran_randm_src.best_score_
```

```python
#gradient boosting classifier
from sklearn.ensemble import GradientBoostingClassifier
parameters = {'loss':['log_loss', 'deviance', 'exponential'],'n_estimators':[10,50,100],'criterion': [ 'friedman_mse',  'mse','squared_error'],'max_depth'    : [1,4,5] ,'random_state': [1,5,3]               }
scoring= ['accuracy']
gb_randm_src = RandomizedSearchCV(estimator=GradientBoostingClassifier(), param_distributions = parameters,scoring='accuracy',cv = 2, n_iter = 10, n_jobs=-1,verbose=1)
gb_randm_src.fit(X_train, y_train)

gradient_best= gb_randm_src.best_score_
```

```python
#xgboost classifier
from xgboost import XGBClassifier
parameters = {'loss':['log_loss', 'deviance', 'exponential'],'n_estimators':[10,50,100],'criterion': [ 'friedman_mse',  'mse','squared_error'],'max_depth'    : [1,4,5] ,'random_state': [1,5,3]                }
scoring= ['accuracy']
xg_randm_src = RandomizedSearchCV(estimator=XGBClassifier(), param_distributions = parameters,scoring='accuracy',cv = 2, n_iter = 10, n_jobs=-1,verbose=1)
xg_randm_src.fit(X_train, y_train)

XGB_best= xg_randm_src.best_score_
```

```python
from sklearn.neighbors import KNeighborsClassifier
parameters = {'n_neighbors' : [20,5,10],'weights':['uniform', 'distance'] ,'algorithm':['auto', 'ball_tree', 'kd_tree','brute']}
clf = KNeighborsClassifier()
knn_randm_src = RandomizedSearchCV(estimator=clf, param_distributions = parameters,scoring='accuracy',cv = 2, n_iter = 10, n_jobs=-1,verbose=1)

knn_randm_src .fit(X_train, y_train)
knn_best= knn_randm_src.best_score_
```

```python
df = {'method':['logistic_regression','decisiontree_classifier','randomforest_classifier','gradientboost_classifier','XGB_classifier','knn_clasifier'],'Best_accuracy_using_randomized_mtd':[log_best,dec_best,random_best,gradient_best,XGB_best,knn_best]}
df1 = pd.DataFrame(df)
  
# displaying the DataFrame
print('DataFrame:\n', df1)
```
