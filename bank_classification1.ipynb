```python
features_na = [features for features in df.columns if df[features].isnull().sum() > 0]
for feature in features_na:
    print(feature,np.round(df[feature].isnull().mean(),4),'%missing values')
else:
    print('no missing value found')
```

```python
for column in df.columns:
    print(column,df[column].nunique())
```

```python
bool_columns =['housing','loan','deposit','default']   
for col in bool_columns:
    df[col]=df[col].apply(lambda x : 1 if x == 'yes' else 0)
df
```

```python
month_mapping = {'jan': 1, 'feb': 2, 'mar': 3,'apr':4,'may' :5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}
df['month'] = df['month'].map(month_mapping)
marital_mapping = {'single' : 0 , 'divorced' : 1 ,'married' :2}
df['marital'] = df['marital'].map(marital_mapping)
education_mapping ={ 'tertiary' :3, 'secondary':2,'primary':1, 'unknown':0}
df['education'] = df['education'].map(education_mapping)
contact_mapping ={'unknown' :0, 'cellular':1, 'telephone':2 }
df['contact'] = df['contact'].map(contact_mapping)
poutcome_mapping ={ 'failure':0, 'success':1,'unknown':2, 'other':3}
df['poutcome'] = df['poutcome'].map(poutcome_mapping)
job_mapping ={'admin.' : 0, 'technician':1, 'services':2, 'management':3, 'retired':4, 'blue-collar':5, 'unemployed':6, 'entrepreneur':7, 'housemaid':8, 'unknown':9, 'self-employed':10,'student':11}
df['job'] = df['job'].map(job_mapping)
df
```

```python
df['balance'].describe()
```

```python
df['balance'] = df['balance'].mask(df['balance'].lt(0)).ffill().fillna(df['balance'].mean()).astype('int32')
df
```

```python
df['balance'].describe()
```

```python
X=df.drop(['deposit'],axis=1).values
y=df['deposit'].values
y
```

```python
#Dividing data in test and train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=50)
```

```python

from sklearn.model_selection import RandomizedSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
from sklearn import linear_model
reg = linear_model.LogisticRegression()
reg.fit(X_train,y_train)
```

```python
log_param= {'penalty':['l1', 'l2', 'elasticnet', 'none'],'dual':[0,1],'solver':['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']}
randm_src = RandomizedSearchCV(estimator=linear_model.LogisticRegression(), param_distributions = log_param,cv = 2, n_iter = 10, n_jobs=-1,verbose=1)
randm_src.fit(X_train, y_train)
log_best= randm_src.best_score_
```

```python
log_score= reg.score(X_test,y_test)
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
clf = tree.DecisionTreeClassifier(max_depth= 4,random_state=1,criterion=   'gini')
clf.fit(X_train, y_train)
clf.score(X_train,y_train)
```

```python

from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
parameters = {'criterion': [ 'gini', 'entropy', 'log_loss'],'max_depth'    : [1,4,5] ,'random_state': [1,5,3]               }
randm_src = RandomizedSearchCV(estimator=model, param_distributions = parameters,cv = 2, n_iter = 10, n_jobs=-1,verbose=1)
randm_src.fit(X_train, y_train)
dec_best=  randm_src.best_score_
```

```python
from sklearn.ensemble import RandomForestClassifier
regr = RandomForestClassifier(random_state=0, n_jobs=-1)
model = regr.fit(X, y)
regr.score(X_test,y_test)
```

```python
parameters = {'n_estimators':[10,50,100],'criterion': [ 'gini', 'entropy', 'log_loss'],'max_depth'    : [1,4,5] ,'random_state': [0,5,3] ,'max_features':['sqrt', 'log2', None]}            

randm_src = RandomizedSearchCV(estimator=RandomForestClassifier(), param_distributions = parameters,cv = 2, n_iter = 10, n_jobs=-1,verbose=1)
randm_src.fit(X_train, y_train)

random_best= randm_src.best_score_
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python

gbr = GradientBoostingClassifier(n_estimators = 500, max_depth = 8 ,min_samples_split=2,min_samples_leaf=2,learning_rate=0.01,loss='log_loss')

# Fit to training set
gbr.fit(X_train, y_train)
```

```python
gbr.score(X_test, y_test)
```

```python
parameters = {'loss':['log_loss', 'deviance', 'exponential'],'n_estimators':[10,50,100],'criterion': [ 'friedman_mse',  'mse','squared_error'],'max_depth'    : [1,4,5] ,'random_state': [1,5,3]               }

randm_src = RandomizedSearchCV(estimator=GradientBoostingClassifier(), param_distributions = parameters,cv = 2, n_iter = 10, n_jobs=-1,verbose=1)
randm_src.fit(X_train, y_train)

gradient_best= randm_src.best_score_
```

```python
parameters = {'loss':['log_loss', 'deviance', 'exponential'],'n_estimators':[10,50,100],'criterion': [ 'friedman_mse',  'mse','squared_error'],'max_depth'    : [1,4,5] ,'random_state': [1,5,3]                }

randm_src = RandomizedSearchCV(estimator=XGBClassifier(), param_distributions = parameters,cv = 2, n_iter = 10, n_jobs=-1,verbose=1)
randm_src.fit(X_train, y_train)

XGB_best= randm_src.best_score_
```

```python
df = {'method':['logistic_regression','decisiontree_classifier','randomforest_classifier','gradientboost_classifier','XGB_classifier'],'score':[ log_score,clf.score(X_train,y_train),regr.score(X_test,y_test),gbr.score(X_test, y_test),xgb_model.score(X_test,y_test)],'Best_score_using_randomized_mtd':[log_best,dec_best,random_best,gradient_best,XGB_best]}
df1 = pd.DataFrame(df)
  
# displaying the DataFrame
print('DataFrame:\n', df1)
```
